```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#plt.style.use('ggplot')
sns.set_palette("husl")

df = pd.read_csv('train.csv')
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df.sort_index(inplace=True)

# Extract time components for our seasonal analysis
df['year'] = df.index.year
df['month'] = df.index.month
df['day_of_week'] = df.index.dayofweek # 0 = Monday, 6 = Sunday

print("Dataset Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(15, 6))

# Resample data to a monthly frequency and sum the sales
monthly_sales = df['sales'].resample('ME').sum()

plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linewidth=2)
plt.title('Overall Monthly Sales Trend (All Stores & Items)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.grid(True, alpha=0.5)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 6))

# Monthly Seasonality
sns.boxplot(data=df, x='month', y='sales', ax=axes[0], showfliers=False)
axes[0].set_title('Sales Distribution by Month (Seasonality)', fontsize=14)
axes[0].set_xlabel('Month (1=Jan, 12=Dec)')
axes[0].set_ylabel('Daily Sales')

# Day of Week Seasonality
sns.boxplot(data=df, x='day_of_week', y='sales', ax=axes[1])
axes[1].set_title('Sales Distribution by Day of Week', fontsize=14)
axes[1].set_xlabel('Day of Week (0=Mon, 6=Sun)')
axes[1].set_ylabel('Daily Sales')

plt.tight_layout()
plt.show()
```
